## train.txt, val.txt 생성

```python
from glob import glob

train_img_list = glob('/content/train/images/*.jpg')
val_img_list = glob('/content/valid/images/*.jpg')
test_img_list = glob('/content/test/images/*.jpg')

train_txt_list = glob('/content/train/labels/*.txt')
val_txt_list = glob('/content/valid/labels/*.txt')
test_txt_list = glob('/content/test/labels/*.txt')

# 파일쓰기
with open('train.txt', 'w') as f:
  f.write('\n'.join(train_img_list)+'\n')

with open('val.txt', 'w') as f:
  f.write('\n'.join(val_img_list)+'\n')

with open('test.txt', 'w') as f:
  f.write('\n'.join(test_txt_list)+'\n')
```

## data.yaml 수정

```python
import yaml

with open('/content/data.yaml', 'r') as f:
  data = yaml.full_load(f)

data['names'] = ['crosswalk', 'green', 'red']
data['train'] = '/content/train.txt'
data['val'] = '/content/val.txt'
data['test'] = '/content/test.txt'

with open('/content/data.yaml', 'w') as f:
  data = yaml.dump(data, f)
```
